```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
pd.set_option('display.max_rows',20)
```

```python
# Styling Charts
plt.style.use('ggplot')
```

```python
surat_data = pd.read_csv("PTax_Demand_Vs_Recovery_2017-18_Wardwise.csv") #,usecols=['Zone_Name', 'Residential_Collection_in_Cr'])
surat_data
```

```python
# Pie chart for Residental collection 2017-18
limbayat = surat_data[surat_data.Zone_Name == 'Limbayat']
limbayat

plt.title('Pie chart for Residental collection 2017-18')
plt.bar(limbayat.Ward_Name,limbayat.Residential_Demand_in_Cr)
plt.xlabel('WardName')
plt.xlim(0,4)

plt.ylabel('Collections in Crores')
plt.bar(limbayat.Ward_Name,limbayat.Residential_Collection_in_Cr)
plt.legend(['Residental 2017-18', 'Resi_Demand 2017-18'])
plt.show()
```

```python
# Total Number of zone available.
plt.title('Total Number of zone available')
plt.ylabel('Zone Name')
plt.xlabel('Total Numbers')
print(surat_data['Zone_Name'].value_counts())
surat_data['Zone_Name'].value_counts().sort_index().plot(kind='bar')
```

```python
# Total Residential Collection according to Zone Name
zone_name = set(surat_data['Zone_Name'])
zone_name = list(zone_name)
rsum = []
for zone in zone_name:
    temp = 0
    data = surat_data[surat_data['Zone_Name']==zone]
    rsum.append(data['Residential_Collection_in_Cr'].sum())

print(zone_name)
print(rsum)

fig, ax = plt.subplots()

ax.barh(zone_name,rsum)
plt.title('Total Residential Collection according to Zone in Surat 2017')
plt.xlabel('Total Sum in Crores')
plt.ylabel('Zone Name')
plt.show()

```

```python
# Total Commercial Collection according to Zone Name
zone_name = set(surat_data['Zone_Name'])
zone_name = list(zone_name)
csum = []
for zone in zone_name:
    temp = 0
    data = surat_data[surat_data['Zone_Name']==zone]
    csum.append(data['Commercial_Collection_in_Cr'].sum())

print(zone_name)
print(csum)

fig, ax = plt.subplots()

ax.barh(zone_name,rsum)
plt.title('Total Commercial Collection according to Zone in Surat 2017')
plt.xlabel('Total Sum in Crores')
plt.ylabel('Zone Name')
plt.show()

```

```python
# Total Industrial Collection according to Zone Name
zone_name = set(surat_data['Zone_Name'])
zone_name = list(zone_name)
isum = []
for zone in zone_name:
    temp = 0
    data = surat_data[surat_data['Zone_Name']==zone]
    isum.append(data['Industrial_Collection_in_Cr'].sum())

print(zone_name)
print(isum)

fig, ax = plt.subplots()

ax.barh(zone_name,rsum)
plt.title('Total Industrial Collection according to Zone in Surat 2017')
plt.xlabel('Total Sum in Crores')
plt.ylabel('Zone Name')
plt.show()

```
