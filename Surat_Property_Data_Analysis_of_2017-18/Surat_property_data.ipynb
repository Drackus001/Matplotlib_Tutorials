```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
pd.set_option('display.max_rows',20)
```

```python
surat_data = pd.read_csv("PTax_Demand_Vs_Recovery_2017-18_Wardwise.csv") #,usecols=['Zone_Name', 'Residential_Collection_in_Cr'])
surat_data
```

```python
# Pie chart for Residental collection 2017-18
limbayat = surat_data[surat_data.Zone_Name == 'Limbayat']
limbayat

plt.title('Pie chart for Residental collection 2017-18')
plt.bar(limbayat.Ward_Name,limbayat.Residential_Demand_in_Cr)
plt.xlabel('WardName')
plt.xlim(0,4)

plt.ylabel('Collections in Crores')
plt.bar(limbayat.Ward_Name,limbayat.Residential_Collection_in_Cr)
plt.legend(['Residental 2017-18', 'Resi_Demand 2017-18'])
plt.show()
```
