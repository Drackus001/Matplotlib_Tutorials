```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 20)
%matplotlib inline
```

```python
surat_data = pd.read_csv('PTax_Demand_Vs_Recovery_2017-18_Wardwise.csv')
surat_data
```

```python
west = 'West'
west_data = surat_data[surat_data['Zone_Name'] == west]
west_data

Wardname = west_data.Ward_Name
Rcollection = west_data.Residential_Collection_in_Cr
```

```python
# Getting started
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots()
ax.barh(Wardname, Rcollection)
```
