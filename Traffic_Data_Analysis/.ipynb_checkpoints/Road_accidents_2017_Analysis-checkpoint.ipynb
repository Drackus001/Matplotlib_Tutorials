```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_rows',15)
```

```python
road_accident_data = pd.read_csv('Road_Accidents_2017-Annuxure_Tables_21.csv')
road_accident_data
```

```python
# Pie chart for total road accidents 2017
states = road_accident_data.States_UTs
states
```

```python
# Road_accidents_at_T-Junction_2017_in_INDIA
plt.style.use('ggplot') # for styling the graph
fig, ax = plt.subplots()
plt.title('Road_accidents_at_T-Junction_2017_in_INDIA')
plt.xlabel('Accidents Numbers')
plt.ylabel('States')
ax.barh(states[0:15], road_accident_data.T_Junction_Total_number_of_Accidents[0:15])
labels = ax.get_xticklabels()
```

```python
# T_Junction_Persons_Killed_2017_in_INDIA
print(plt.style.available)# for styling the graph
plt.style.use('seaborn')
fig, ax = plt.subplots()
plt.title('T_Junction_Persons_Killed_2017_in_INDIA')
plt.xlabel('Killed in Numbers')
plt.ylabel('States')
ax.barh(states[0:15], road_accident_data.T_Junction_Persons_Killed[0:15])
#labels = ax.get_xticklabels()
```

```python
# Pie chart for Staggered_Junction_Persons_Killed_in_2017
killed_number__in_staggered = road_accident_data.Staggered_Junction_Persons_Killed
killed_number__in_staggered

fig, ax = plt.subplots()
ex = (0,0,0,0,0.1) # 0.1 will be sliced-out
#print(killed_number__in_staggered[0:5])
#print(states[0:5])
plt.title('Staggered_Junction_Persons_Killed_in_2017')
ax.pie(killed_number__in_staggered[2:7], explode=ex, labels=states[2:7],  autopct='%1.1f%%',
       shadow=True, startangle=90)
ax.axis('equal')

plt.show()
```
